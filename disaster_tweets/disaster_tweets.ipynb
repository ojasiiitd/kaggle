```python
import pandas as pd
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
text = [sent for sent in train.text]
outcome = train.target
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(text , outcome)
vect = TfidfVectorizer(max_features=1600 , min_df=0.001 , max_df=0.85 , ngram_range=(1,2))
x_train_vect = vect.fit_transform(x_train)
```

```python
vect.get_feature_names()
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold, cross_val_score
```

```python
clf = RandomForestClassifier()
clf.fit(x_train_vect , y_train)

x_test_vect = vect.transform(x_test)
clf.score(x_test_vect , y_test)
```

```python
clf = MultinomialNB()
clf.fit(x_train_vect , y_train)

x_test_vect = vect.transform(x_test)
clf.score(x_test_vect , y_test)
```

```python
grid_params = {"C":[10 , 100] , "gamma":[0.001 , 0.01]}
clf = SVC()
grid = GridSearchCV(clf , grid_params , n_jobs=-1 , verbose=2)
grid.fit(x_train_vect , y_train)
```

```python
grid.best_estimator_ , grid.best_score_
```

```python
clf = SVC(C=100 , gamma=0.01)
clf.fit(x_train_vect , y_train)

x_test_vect = vect.transform(x_test)
clf.score(x_test_vect , y_test)
```

```python
vect = TfidfVectorizer(max_features=1600 , min_df=0.001 , max_df=0.85 , ngram_range=(1,2))
train_vect = vect.fit_transform(text)
outcome
```

```python
grid_params = {"C":[10 , 100] , "gamma":[0.001 , 0.01]}
clf = SVC()
grid = GridSearchCV(clf , grid_params , n_jobs=-1 , verbose=2)
grid.fit(train_vect , outcome)
grid.best_estimator_
```

```python
clf = SVC(C=10 , gamma=0.01)
clf.fit(train_vect , outcome)
```

```python
testText = [sent for sent in test.text]
x_test = vect.transform(testText)
```

```python
soln = pd.DataFrame()
predicted = clf.predict(x_test)
```

```python
soln["id"] = test.id
soln["target"] = predicted
print(soln)
```
