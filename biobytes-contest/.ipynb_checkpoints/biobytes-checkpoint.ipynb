```python
num = 1
map = {}
for d in data[0] :
    if d[0] == ">":
        key = d[1:-2]
        if key in map :
            continue
        map[key] = num
        num += 1
map
```

```python
protein = []
alphabet = []
bind = []
for i in range(len(data[0])) :
    if data[0][i][0] == ">":
        key = data[0][i][1:-2]
        for j in range(len(data[0][i+1])) :
            protein.append(map[key])
            alphabet.append(ord(data[0][i+1][j]))
            bind.append(data[0][i+2][j])
```

```python
x = pd.DataFrame()
x["protein"] = protein
x["alpha"] = alphabet
y = bind
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV , GridSearchCV , train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
clf = SVC()
clf.fit(x_train_scaled , y_train)
```

```python
x_test_scaled = scaler.transform(x_test)
clf.score(x_test_scaled , y_test)
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
params = {
    "kernel": ["linear" , "poly" , "rbf" , "sigmoid" , "precomputed"],
    "gamma": [10 , 100 , 500 , 1000],
    "C": [0.001 , 0.01 , 0.1]
}
tempclf = RandomizedSearchCV(SVC() , params , n_jobs=-1)
tempclf.fit(x_scaled , y)
```

```python
print(tempclf.best_estimator_ , tempclf.best_score_)
```
